## IPYTHON

- 1. Comandos:

python3 -m venv env -> nombre del entorno cualquiera

ls -> Para verlo

source env/bin/activate -> para linux

.\env\scripts\activate -> windows

pip install ipython

- 2. Para ingresar a la consola CMD:

ipython

tab -> autocompletar names variables

Ejemplo -> full_names.upper() "mayuscula"

exit -> para salir

- 3. Comandos magicos

%pwd -> directorio donde me encuentro

%cd -> cambiar directorio

%ls -> conocer archivos o carpetas

%bookmark projects -> almacenar un directorio en una variable

%bookmark -l -> listar los directorios

import os -> os.chdir("ruta")

%cd -b direct -> para cambiar de ruta

cody="cody" -> %store cody "almacenar variables"

%store -r -> cargar todas las variables almacenadas

%lsmagic -> ver todos los comandos magicos disponibles

- 4. Instronspeccion del codigo:

def suma(a=int,b=int))  -> int:

"permite sumar dos numeros enteros"

return a+b

print(suma(10,20))

suma(10,20)

suma? -> resumen del objeto

suma?? -> para ver el codigo de la funcion

- 5. Manipulacion de archivos:

code . -> desde la ruta

%pycat ejemplo1.py -> conocer el contenido

%run ejemplo1.py -> ejecutar el codigo

%load ejemplo1.py -> preview, confirma codigo a cargar y editable

%pdoc suma -> documentacion de una funcion

ipython -i ejemplo1.py -> cargar el mismo resultado en el shell y se puede usar todo lo que contiene

%edit ejemplo1.py -> se abre el bloc de notas que se puede editar y guardar
